```python
#Import Thread_benchmarking.csv file
#Use conda env Test_V2_ITSxpress200
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.use('TkAgg')


df = pd.read_csv('ThreadsBenchmarking_exclusivetag.csv')
df.head()
```

```python
#Add a column to the dataframe that calculates the standard deviation of the each thread for each region for the column V2_1out.gz_Time_elapsed
df['V2_1out.gz_Time_elapsed_std'] = df.groupby(['region','threads'])['V2_1out.gz_Time_elapsed'].transform('std')
#Add a column to the dataframe that calculates the average of the each thread for each region for the column V2_1out.gz_Time_elapsed
df['V2_1out.gz_Time_elapsed_mean'] = df.groupby(['region','threads'])['V2_1out.gz_Time_elapsed'].transform('mean')
#Add a column to the dataframe that calculates the standard deviation of the each thread for each region for the column V1_1out.gz_Time_elapsed
df['V1_1out.gz_Time_elapsed_std'] = df.groupby(['region','threads'])['V1_1out.gz_Time_elapsed'].transform('std')
#Add a column to the dataframe that calculates the average of the each thread for each region for the column V1_1out.gz_Time_elapsed
df['V1_1out.gz_Time_elapsed_mean'] = df.groupby(['region','threads'])['V1_1out.gz_Time_elapsed'].transform('mean')
#Add a column to the dataframe that calculates the standard deviation of the each thread for each region for the column ITSx_time_elapsed
df['ITSx_time_elapsed_std'] = df.groupby(['region','threads'])['ITSx_time_elapsed'].transform('std')
#Add a column to the dataframe that calculates the average of the each thread for each region for the column ITSx_time_elapsed
df['ITSx_time_elapsed_mean'] = df.groupby(['region','threads'])['ITSx_time_elapsed'].transform('mean')
df.to_csv('ThreadsBenchmarking_added_means.csv', index=False)
#Create new dataframe called df_filtered with the following columns: region, threads, V2_1out.gz_Time_elapsed_std, V2_1out.gz_Time_elapsed_mean, V1_1out.gz_Time_elapsed_std, V1_1out.gz_Time_elapsed_mean, ITSx_time_elapsed_std, ITSx_time_elapsed_mean
df_filtered = df[['region','threads','V2_1out.gz_Time_elapsed_std','V2_1out.gz_Time_elapsed_mean','V1_1out.gz_Time_elapsed_std','V1_1out.gz_Time_elapsed_mean','ITSx_time_elapsed_std','ITSx_time_elapsed_mean']].drop_duplicates()
df_filtered.head()
df_filtered.to_csv('ThreadsBenchmarking_exclusivetag_filtered.csv', index=False)
```
